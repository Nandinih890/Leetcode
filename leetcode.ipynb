```python
nums = list(map(int, input().split()))
target = int(input())

lookup = {}

for i, num in enumerate(nums):
    complement = target - num

    if complement in lookup:
        print([lookup[complement], i])
        break

    lookup[num] = i
```

```python
nums = list(map(int, input().split()))

low = 0
mid = 0
high = len(nums) - 1

while mid <= high:
    if nums[mid] == 0:
        nums[low], nums[mid] = nums[mid], nums[low]
        low += 1
        mid += 1

    elif nums[mid] == 1:
        mid += 1

    else:
        nums[mid], nums[high] = nums[high], nums[mid]
        high -= 1

print(*nums)
```
